```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plddt_antifam = pd.read_csv('pLDDT_seq_length_info_AntiFam.csv')
ptm_antifam = pd.read_csv('pTM_seq_length_info_AntiFam.csv')

plddt_antifam = plddt_antifam[plddt_antifam['Sequence length'] <= 200]
ptm_antifam = ptm_antifam[ptm_antifam['Sequence length'] <= 200]

plddt_swissprot = pd.read_csv('pLDDT_seq_length_info_SwissProt.csv')
ptm_swissprot = pd.read_csv('pTM_seq_length_info_SwissProt.csv')
```

```python
# pLDDT
fig, ax = plt.subplots(figsize=(10, 6))
                                                                                                                                                                                                                                        
ax.scatter(plddt_swissprot['Sequence length'], plddt_swissprot['average pLDDT'], 
           c='#bf525d', label='SwissProt')
ax.scatter(plddt_antifam['Sequence length'], plddt_antifam['plDDT mean'], 
           c='#3976af', label='AntiFam')

ax.legend()
plt.xlabel('Sequence length')
plt.ylabel('pLDDT')
                                                                                                                                                                                                                                           
plt.show()
                   
```

```python
# pTM
fig, ax = plt.subplots(figsize=(10, 6))
                                                                                                                                                                                                                                        
ax.scatter(ptm_swissprot['Sequence length'], ptm_swissprot['pTM'], 
           c='#bf525d', label='SwissProt')
ax.scatter(ptm_antifam['Sequence length'], ptm_antifam['pTM'], 
           c='#3976af', label='AntiFam')

ax.legend()
plt.xlabel('Sequence length')
plt.ylabel('pTM')
                                                                                                                                                                                                                                           
plt.show()
```
